```python
df = df.dropna()
```

#**PHASE C: Handle Class Imbalance with SMOTE**

**Step 1: Split Train and Test Sets**

```python
from sklearn.model_selection import train_test_split

# Split before applying SMOTE (to avoid data leakage)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Check class distribution in train set
print("Before SMOTE:")
print(y_train.value_counts())
```

**Step 2: Apply SMOTE to Balance the Training Set**

**Downsample + Redefine X and y**

```python
# Downsample to 5000 rows
df_sampled = df_encoded.sample(n=5000, random_state=42)

# Redefine features and target
X = df_sampled.drop('claim_status', axis=1)
y = df_sampled['claim_status']

# Confirm new shape
print("X shape:", X.shape)
print("y shape:", y.shape)
```

**Train-Test Split (Again)**


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Before SMOTE:")
print(y_train.value_counts())
```

**Apply SMOTE on Downsampled Data**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("After SMOTE:")
print(y_train_resampled.value_counts())
```

#**PHASE D: Train & Evaluate XGBoost Model**

**Step 1: Train the Model**

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

# Train XGBoost
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train_resampled, y_train_resampled)
```

**Step 2: Make Predictions on the Test Set**

```python
# Predict on original test set (NOT resampled)
y_pred = xgb_model.predict(X_test)
```

**Step 3: Evaluate Performance**

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# AUC Score
print("AUC Score:", roc_auc_score(y_test, y_pred))
```

**Random Forest Classifier**

**Train Random Forest Model**

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)

# Predict
rf_pred = rf_model.predict(X_test)
```

**Evaluate Random Forest**

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print("Random Forest Results:")
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_pred))
print("Classification Report:\n", classification_report(y_test, rf_pred))
print("AUC Score:", roc_auc_score(y_test, rf_pred))
```

**Comparison Code: Accuracy, Precision, Recall, F1, AUC**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# XGBoost Predictions
xgb_pred = xgb_model.predict(X_test)

# Random Forest Predictions (already done earlier)
rf_pred = rf_model.predict(X_test)

# Initialize comparison dictionary
metrics = {
    'Model': ['XGBoost', 'RandomForest'],
    'Accuracy': [
        accuracy_score(y_test, xgb_pred),
        accuracy_score(y_test, rf_pred)
    ],
    'Precision': [
        precision_score(y_test, xgb_pred),
        precision_score(y_test, rf_pred)
    ],
    'Recall': [
        recall_score(y_test, xgb_pred),
        recall_score(y_test, rf_pred)
    ],
    'F1 Score': [
        f1_score(y_test, xgb_pred),
        f1_score(y_test, rf_pred)
    ],
    'AUC': [
        roc_auc_score(y_test, xgb_pred),
        roc_auc_score(y_test, rf_pred)
    ]
}

# Create DataFrame for easy viewing
comparison_df = pd.DataFrame(metrics)
comparison_df.set_index('Model', inplace=True)

# Display result
comparison_df
```

#**Hyper Paramter Tuning**

**1. Add scale_pos_weight to handle imbalance**

This helps XGBoost treat minority class (1 = risky claim) more seriously.

```python
# Compute the ratio of class 0 to class 1 in training data
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

xgb_model = XGBClassifier(
    eval_metric='logloss',
    scale_pos_weight=scale_pos_weight,
    random_state=42
)
xgb_model.fit(X_train_resampled, y_train_resampled)
```

**2. Use GridSearchCV to find best hyperparameters**

You use GridSearch after confirming a model works

Use it before final training, to avoid overfitting and underfitting

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Already computed earlier
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

# Light grid
param_grid = {
    'n_estimators': [100],
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1],
    'scale_pos_weight': [scale_pos_weight]
}

xgb = XGBClassifier(eval_metric='logloss', random_state=42)

grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='f1',
    cv=2,               # reduced from 3 to 2
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train_resampled, y_train_resampled)
print("Best Params:", grid_search.best_params_)

best_model = grid_search.best_estimator_
```

**3. Evaluate the Best Model**

```python
y_pred = best_model.predict(X_test)

from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, y_pred))
```

**4. Check for Overfitting or Underfitting**

Use cross-validation or plot learning curves:

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(best_model, X_train_resampled, y_train_resampled, cv=5, scoring='f1')
print("Cross-validation F1-scores:", scores)
print("Mean F1:", scores.mean())
```

#**PHASE 3B: GenAI-Based Explanation Using Hugging Face**

Now we’ll use SHAP output to generate natural language explanations like:

“This claim is classified as high-risk because the customer is from region X, owns vehicle type Y, and the claim amount is Z.”

**Step 1: Install Transformers**

```python
!pip install transformers --quiet
```

**Step 2: Load a Small GenAI Model (T5 or GPT2)**

```python
from transformers import pipeline

# Use a small, fast text generation model
genai = pipeline("text-generation", model="distilgpt2")
```

**Step 3: Format SHAP Output → Prompt Template**

Create a claim explanation prompt:

```python
# Simulate top features from SHAP
features_used = X_test.iloc[index][shap_values[index].values > 0].to_dict()

# Convert to a natural language prompt
prompt = "Explain why this claim is high-risk based on the following: "
for k, v in features_used.items():
    prompt += f"{k} is {v}, "

print("Prompt:", prompt)
```

**Step 4: Generate Explanation with GenAI**

```python
explanation = genai(prompt, max_length=50, do_sample=True, temperature=0.7)
print("GenAI Explanation:", explanation[0]['generated_text'])
```

**Improved Prompt Format Example:**

```python
top_features = X_test.iloc[index][shap_values[index].values > 0].sort_values(ascending=False)[:5]
prompt = "Explain why this claim is high-risk based on these conditions:\n"

for feature, value in top_features.items():
    prompt += f"- {feature.replace('_', ' ')} is {value}\n"

print("Prompt for GenAI:\n", prompt)
```

```python
explanation = genai(prompt, max_new_tokens=50, do_sample=True, temperature=0.7)
print("GenAI Explanation:", explanation[0]['generated_text'])
```
